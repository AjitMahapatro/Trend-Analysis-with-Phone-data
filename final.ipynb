# Data Transformation

```python
 #Calculating mean battery drain for each device model and store it in a column
DS['mean_battery_drain']= DS.groupby('Device Model')['Battery Drain (mAh/day)'].transform('mean')

 #Calculating mean data usage for each device model and store it in another column
DS['mean_data_usage']= DS.groupby('Device Model')['Data Usage (MB/day)'].transform('mean')

#Displaying the updated DataFrame
DS
```

# Feature Engineering

```python
 #Creating a new feature: Battery Efficiency
DS['Battery Efficiency']= DS['Battery Drain (mAh/day)']/DS['Screen On Time (hours/day)']
print(DS[['Battery Drain (mAh/day)','Screen On Time (hours/day)','Battery Efficiency']].head(10))
```

# Statistical Analysis

```python
 #to print maximum values of a column
 #Here, max_perc is a column that shows the percentage battery drain values
DS['max_perc']= (DS['Battery Drain (mAh/day)'] / DS['Battery Drain (mAh/day)'].max())*100
print(DS['max_perc'])
```

```python
 #to print minimum values of a column
 #Here, min_perc is a column that shows the minimum values
DS['min_perc']= (DS['Number of Apps Installed'] / DS['Number of Apps Installed'].min())*100
print(DS['min_perc'])
```

```python
 #Finding mean value of a column(Ex: Data Usage (MB/day))
mean_val= DS['Data Usage (MB/day)'].mean()
print(mean_val)
```

```python
 #Finding median value of a column(Ex: Data Usage (MB/day))
median_val= DS['Data Usage (MB/day)'].median()
print(median_val)
```

```python
 #Finding mode value of a column(Ex: Data Usage (MB/day))
mode_val= DS['Data Usage (MB/day)'].mode()[0]
print(mode_val)
```

```python
 #Applying standard deviation on a specific column
std_val= DS['Data Usage (MB/day)'].std()
print(std_val)
```

```python
 #To print sum of all values in a column
sum_sc= DS['Screen On Time (hours/day)'].sum()
print(sum_sc)
```

```python
  #Minimum value in a column
min_sc= DS['Screen On Time (hours/day)'].min()
print(min_sc)
 
 #Maximum value in a column
max_sc= DS['Screen On Time (hours/day)'].max()
print(max_sc)

  #Range of the column
range_sc= max_sc-min_sc
print(range_sc)
```

```python
 #Identifies unique column values("Female", "Male"), and converts to numerical values.
le=LabelEncoder()
DS['Gender']=le.fit_transform(DS['Gender'])
print(DS['Gender'])
```

```python
 #Filter rows for finding upper outliers
upper= DS[DS['Data Usage (MB/day)'] >(mean_val+2*std_val)]
print(upper)
```

```python
 #Filter rows for finding lower outliers
lower= DS[DS['Data Usage (MB/day)'] <(mean_val-2*std_val)]
print(lower)
```

## Correlation Heatmap

```python
 #selecting columns having numerical values 
num_cols= DS.select_dtypes(include=['int64','float64']).columns

 #Here, we are filtering only integer and floating value columns
print(num_cols)
```

```python
#creating a correlation matrix of dataset "DS"
correlation_matrix= DS[num_cols].corr()
print(correlation_matrix)
```

```python
#Heatmap visualization 
  #setting figure cube size
plt.figure(figsize=(8,5))
  #mapping correlation with the correlation matrix of Dataset
sb.heatmap(correlation_matrix,annot=True,cmap='Blues')

plt.title("Correlation Heatmap",fontweight='bold',pad=10)
  #to assign x-axis labels as the column names
plt.xticks(rotation=45, ha='right')
  #display the heatmap
plt.show()
```

## Pie-Chart

```python
 #to find the number of users using a particular mobile model using value_counts()
user_count= DS['Device Model'].value_counts()
print(user_count)
```

```python
 #Pie-Chart for representing the user count of Mobile Models
colr= plt.cm.Blues([50,100,150,200,250]) 
plt.pie(user_count, labels=DS['Device Model'].value_counts().index,colors=colr,autopct='%1.1f%%')
plt.title('Pie Sections of User')
plt.show()
```

## 3D Scatter Plot

```python
 #Creating figure for 3D Scatter Plot
fig= plt.figure(figsize=(6,6))
 #Adding subplots
sub= fig.add_subplot(111, projection='3d')
 #Giving 3 column values for 3 Dimensional plot
sub.scatter(DS['Number of Apps Installed'],DS['Data Usage (MB/day)'],DS['Battery Drain (mAh/day)'])
 #Title of the Plot
sub.set_title('3D Scatter Plot')
sub.set_xlabel('X Axis')
sub.set_ylabel('Y Axis')
sub.set_zlabel('Z Axis')
plt.show()
```

## Donut Chart

```python
 #Grouping the types of device models and aggregating
data=DS.groupby('Device Model')['Data Usage (MB/day)'].sum()
 #customizing colors to the chart
ls=plt.cm.Blues([50,100,150,200,250])
 #creating Donut chart
plt.pie(data,labels=data.index,autopct='%1.1f%%',colors=ls,wedgeprops=dict(width=0.3))
plt.title('Mobile Model VS Data Usage')
plt.show()
```
